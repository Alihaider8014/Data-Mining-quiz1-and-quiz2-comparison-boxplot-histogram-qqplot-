```python
#import the library
import numpy as np
import pandas as pd
import matplotlib as plt
```

```python
#mean, std,min, max, count of the both classes
#and we easily compare the data of class a and class b becasue we have the both classes min, max marks 
#and also mean 
df.describe()
```
